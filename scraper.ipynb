```python
import requests
import pdfplumber
import os
import json
from bs4 import BeautifulSoup
import io


url = "https://www.soa.org/education/exam-req/syllabus-study-materials/edu-multiple-choice-exam/"
resp = requests.get(url)
soup = BeautifulSoup(resp.content, 'html.parser')

link = soup.find_all('a')
links_list = []
count = 0

# Gets all the link
for l in link:
    href = l.get('href')
    if href is not None:
        if "solution" in href and "pdf" in href:
            links_list.append("https://www.soa.org" + href)
            count += 1
print(links_list)
```

```python
data = []
for l in range(len(links_list)):
    resp = requests.get(links_list[l])
    pdf = pdfplumber.open(io.BytesIO(resp.content))
    counter = 0
    for page in pdf.pages: 
        data.append({f'link number:{l}, page number:{counter}': page.extract_text().strip()})
        with open(f"output.json", "w") as f:
            json.dump(data, f, indent=2)
        counter += 1
print("something happened")

```
